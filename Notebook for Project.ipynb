```python
df.sort_values(by="Country", ascending=True, inplace=True)
```

```python
df
```

```python
conditions = [
    df["Age"] < 18,
    df["Age"].between(18, 20),
    df["Age"].between(21, 23),
    df["Age"].between(24, 30),
    df["Age"] > 30]
choices = ["<18", "18-20", "21-23", "24-30", "30+"]
df["Age Group"] = np.select(conditions, choices,default="Unknown")
df
```

```python
conditions = [
    df["Sleep_Hours_Per_Night"] < 5,
    df["Sleep_Hours_Per_Night"] <= 7,
    df["Sleep_Hours_Per_Night"] > 7]
choices = ["Short Sleep", "Normal Sleep", "Long Sleep"]
df["Sleep Category"] = np.select(conditions, choices, default="Unknown")
df
```

```python
df["Addiction Category"] = df["Addicted_Score"].apply(
    lambda x: "Low" if x <= 3 
    else "Moderate" if x <= 6 
    else "High")
df
```

```python
df["Conflict Category"] = df["Conflicts_Over_Social_Media"].apply(
    lambda x: "Low" if x <= 1
    else "Moderate" if x <= 3
    else "High"
)
df
```

```python
df["Usage Group"] = pd.cut(
    df["Avg_Daily_Usage_Hours"],
    bins=[0, 3, 6, float("inf")],
    labels=["Low Users", "Moderate Users", "Heavy Users"],
    right=False
)
df
```

```python
#1 most used platform VS Avg daily usage hours
top_items = df.groupby('Most_Used_Platform')['Avg_Daily_Usage_Hours'].mean().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,5))
colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9', '#92A8D1']
top_items.plot(kind='bar', color=colors, edgecolor='black')
plt.title("Top 5 Platforms")
plt.xlabel("Most Used Platform")  
plt.ylabel("Avg Daily Usage Hours")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#2 Addiction score by ag group 
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Age Group', y='Addicted_Score', hue='Age Group',      palette='coolwarm', errorbar=None, edgecolor='black',legend=False,
order=['18-20', '21-23', '24-30'] )
plt.title("Addiction Score by Age Group", fontsize=14, fontweight='bold')
plt.xlabel("Age Group")
plt.ylabel("Average Addiction Score")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
```

```python
# Objective 3: Examine the relationship between social media addiction and mental health
correlation = df['Addicted_Score'].corr(df['Mental_Health_Score'])
print("Correlation between Addiction and Mental Health:", correlation)
plt.figure(figsize=(8,5))
plt.scatter(df['Addicted_Score'], df['Mental_Health_Score'], color='purple', alpha=0.6, edgecolor='black')
plt.xlabel("Addiction Score")
plt.ylabel("Mental Health Score")
plt.title("Addiction vs Mental Health")
m, b = np.polyfit(df['Addicted_Score'], df['Mental_Health_Score'], 1)
plt.plot(df['Addicted_Score'], m*df['Addicted_Score'] + b, color='red', linewidth=2)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Insight 4: Comparison Between Average Usage Hours and Addiction Levels Across Social Media Platforms
platform_usage = df.groupby('Most_Used_Platform')['Avg_Daily_Usage_Hours'].mean()
platform_addiction = df.groupby('Most_Used_Platform')['Addicted_Score'].mean()
plt.figure(figsize=(10,6))
plt.bar(platform_usage.index, platform_usage.values, color='skyblue', alpha=0.7, label='Avg Daily Usage Hours')
plt.plot(platform_addiction.index, platform_addiction.values, color='red', marker='o', linewidth=2, label='Addiction Score')
plt.title("Average Daily Usage Hours vs Addiction Score by Platform", fontsize=14, fontweight='bold')
plt.xlabel("Social Media Platforms")
plt.ylabel("Average Values")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
```

```python
#Insight 5: Relationship Between Conflict Category and Daily Usage Hours
sns.barplot(data=df, x='Conflict Category', y='Avg_Daily_Usage_Hours',hue='Conflict Category', palette='coolwarm', legend=False)
plt.title('Relationship Between Conflict Category and Daily Usage')
plt.xlabel('Conflict Category')
plt.ylabel('Average Daily Usage (Hours)')
```

```python
# Insight 6: Impact of Social Media Usage on Academic Performance
counts = df['Affects_Academic_Performance'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set2.colors)
plt.title("Effect of Social Media on Academic Performance")
plt.legend(title="Affects Academic Performance", loc="best")
```

```python
# Insight 7: Effect of Relationship Status on Social Media Conflicts
sns.barplot(data=df,x='Relationship_Status',y='Conflicts_Over_Social_Media',hue='Relationship_Status',palette='coolwarm',legend=False)
plt.title("Conflicts by Relationship Status")
plt.xlabel("Relationship Status")
plt.ylabel("Average Conflict Level")
```

```python
# Insight 8: Role of Social Media in Social Relationships (Conflicts by Category)
df['Conflict Category'].value_counts().plot(kind='bar',color='skyblue',edgecolor='black')
plt.title("Levels of Conflicts Caused by Social Media")
plt.xlabel("Conflict Category")
plt.ylabel("Number of Students")
plt.tight_layout()
```

```python
#9 Number of students affected academically by Sleep Category
counts = df.groupby('Sleep Category')['Affects_Academic_Performance'].value_counts().unstack(fill_value=0)
counts.plot(kind='bar', stacked=True, figsize=(8,5), colormap='Set2', edgecolor='black')
plt.title("Impact of Sleep Hours on Academic Performance")
plt.xlabel("Sleep Category")
plt.ylabel("Number of Students")
plt.tight_layout()
```

```python
# 10 :Assess the combined impact of addiction on studentsâ€™ mental health
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='Addiction Category',y='Mental_Health_Score',hue='Addiction Category',palette='coolwarm',edgecolor='black',legend=False)
plt.title("Impact of Social Media Addiction on Mental Health")
plt.xlabel("Addiction Category")
plt.ylabel("Average Mental Health Score")
plt.tight_layout()
```

```python
#Insight 11: Key numeric variables show how addiction, sleep, and mental health are interrelated.
cols = ['Addicted_Score', 'Sleep_Hours_Per_Night', 'Mental_Health_Score']
plt.figure(figsize=(6,5))
sns.heatmap(df[cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Key Numeric Variables")
plt.tight_layout()
```

```python
#Insight 12: Students are grouped by social media usage patterns to identify distinct clusters, such as heavy VS moderate users, and compare their behaviors.
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Avg_Daily_Usage_Hours',y='Addicted_Score',hue='Usage Group',palette='Set1',s=100)
plt.title("Student Clusters by Usage and Addiction")
plt.xlabel("Average Daily Usage Hours")
plt.ylabel("Addiction Score")
plt.tight_layout()
```
